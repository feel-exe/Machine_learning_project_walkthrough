## Imputing Missing Values

Стандартные модели машинного обучения не могут работать с пропущенными значениями, и это означает, что мы должны найти способ их заполнить или отбросить любые функции с пропущенными значениями. Поскольку в первой части мы уже удалили элементы с пропущенными значениями более чем на 50%, здесь мы сосредоточимся на заполнении этих пропущенных значений (процесс, известный как вменение). Существует несколько методов вменения, но здесь мы будем использовать относительно простой метод замены отсутствующих значений медианой столбца. (Вот более подробное обсуждение вменения пропущенных значений)

В приведенном ниже коде мы создаем объект Imputer Scikit-learn, чтобы заполнить пропущенные значения медианой столбца. Обратите внимание, что мы тренируем импутер (используя метод Imputer.fit) на данных обучения, но не на данных тестирования. Затем мы преобразуем (используя Imputer.transform) как данные обучения, так и данные тестирования. Это означает, что пропущенные значения в наборе тестирования заполняются медианным значением соответствующих столбцов в обучающем наборе. Мы должны делать это таким образом, а не потому, что во время развертывания нам придется вменять пропущенные значения в новые наблюдения на основе предыдущих данных обучения. Это один из способов избежать проблемы, известной как утечка данных, когда информация из набора тестирования «просачивается» в процесс обучения.

```python
# Create an imputer object with a median filling strategy
imputer = SimpleImputer(missing_values=np.nan, strategy='mean')

# Train on the training features
imputer.fit(train_features)

# Transform both training data and testing data
X = imputer.transform(train_features)
X_test = imputer.transform(test_features)
```

```python
# Недостающие значения в выборке
print('Missing values in training features: ', np.sum(np.isnan(X)))
print('Missing values in testing features:  ', np.sum(np.isnan(X_test)))
```

```python
# Make sure all values are finite Убедитесь, что все значения конечны
print(np.where(~np.isfinite(X)))
print(np.where(~np.isfinite(X_test)))
```

После вменения все функции являются реальными. Для более сложных методов вменения (хотя медианные значения обычно работают хорошо), ознакомьтесь с этой статьей

###  Особенности масштабирования

Последний шаг, который необходимо предпринять, прежде чем мы сможем построить наши модели, - это масштабирование функций. Это необходимо, потому что объекты находятся в разных единицах, и мы хотим нормализовать функции, чтобы эти единицы не влияли на алгоритм. Линейная регрессия и случайный лес не требуют масштабирования объектов, но другие методы, такие как машины опорных векторов и k ближайших соседей, требуют этого, потому что они учитывают евклидово расстояние между наблюдениями. По этой причине рекомендуется масштабировать функции при сравнении нескольких алгоритмов.

Есть два способа масштабирования функций:

    1 Для каждого значения вычтите среднее значение элемента и поделите на стандартное отклонение элемента. Это называется стандартизацией и приводит к тому, что каждое свойство имеет среднее значение 0 и стандартное отклонение

    2 Для каждого значения вычтите минимальное значение объекта и разделите на максимум минус минимум для объекта (диапазон). Это гарантирует, что все значения для объекта находятся в диапазоне от 0 до 1 и называется масштабированием до диапазона или нормализацией.
Вот хорошая статья о нормализации и стандартизации.

Как и при вменении, когда мы тренируем масштабируемый объект, мы хотим использовать только обучающий набор. Когда мы преобразуем функции, мы преобразуем как обучающий набор, так и набор для тестирования.

```python
# Create the scaler object with a range of 0-1
scaler = MinMaxScaler(feature_range=(0, 1))

# Fit on the training data
scaler.fit(X)

# Transform both the training and testing data
X = scaler.transform(X)
X_test = scaler.transform(X_test)
```

```python

# Convert y to one-dimensional array (vector) Преобразовать y в одномерный массив (вектор)
y = np.array(train_labels).reshape((-1, ))
y_test = np.array(test_labels).reshape((-1, ))
```

Для сравнения моделей мы будем в основном использовать значения по умолчанию Scikit-Learn для гиперпараметров модели. Вот что говорит Scikit-Learn документация по умолчанию:

__Sensible defaults__: всякий раз, когда операции требуется пользовательский параметр,
соответствующее значение по умолчанию определяется библиотекой. Значение по умолчанию

решение поставленной задачи.)

Одна из лучших частей scikit-learn заключается в том, что все модели реализованы одинаковым образом: если вы знаете, как их создать, вы можете реализовать чрезвычайно разнообразный массив моделей. Здесь мы реализуем все процедуры обучения и тестирования для ряда моделей всего за несколько строк кода.

```python
# Function to calculate mean absolute error
def mae(y_true, y_pred):
    return np.mean(abs(y_true - y_pred))

# Takes in a model, trains the model, and evaluates the model on the test set
def fit_and_evaluate(model):
    
    # Train the model
    model.fit(X, y)
    
    # Make predictions and evalute
    model_pred = model.predict(X_test)
    model_mae = mae(y_test, model_pred)
    
    # Return the performance metric
    return model_mae
```

```python
lr = LinearRegression()
lr_mae = fit_and_evaluate(lr)

print('Linear Regression Performance on the test set: MAE = %0.4f' % lr_mae)
```

```python
svm = SVR(C = 1000, gamma = 0.1)
svm_mae = fit_and_evaluate(svm)

print('Support Vector Machine Regression Performance on the test set: MAE = %0.4f' % svm_mae)
```

```python
random_forest = RandomForestRegressor(random_state=60)
random_forest_mae = fit_and_evaluate(random_forest)

print('Random Forest Regression Performance on the test set: MAE = %0.4f' % random_forest_mae)
```

```python
gradient_boosted = GradientBoostingRegressor(random_state=60)
gradient_boosted_mae = fit_and_evaluate(gradient_boosted)

print('Gradient Boosted Regression Performance on the test set: MAE = %0.4f' % gradient_boosted_mae)
```

```python
knn = KNeighborsRegressor(n_neighbors=10)
knn_mae = fit_and_evaluate(knn)

print('K-Nearest Neighbors Regression Performance on the test set: MAE = %0.4f' % knn_mae)
```

```python
plt.style.use('fivethirtyeight')
figsize(8, 6)

# Dataframe to hold the results
model_comparison = pd.DataFrame({'model': ['Linear Regression', 'Support Vector Machine',
                                           'Random Forest', 'Gradient Boosted',
                                            'K-Nearest Neighbors'],
                                 'mae': [lr_mae, svm_mae, random_forest_mae, 
                                         gradient_boosted_mae, knn_mae]})

# Horizontal bar chart of test mae
model_comparison.sort_values('mae', ascending = False).plot(x = 'model', y = 'mae', kind = 'barh',
                                                           color = 'lightgreen', edgecolor = 'black')

# Plot formatting
plt.ylabel(''); plt.yticks(size = 14); plt.xlabel('Mean Absolute Error'); plt.xticks(size = 14)
plt.title('Model Comparison on Test MAE', size = 20);
```

В зависимости от пробега (точные результаты меняются каждый раз), регрессор повышения градиента работает лучше всего, за которым следует случайный лес. # Model Optimization
В машинном обучении оптимизация модели означает поиск наилучшего набора гиперпараметров для конкретной задачи.

### Гиперпараметры
Прежде всего, нам нужно понять, какие гиперпараметры модели отличаются от параметров модели:

Гиперпараметры модели лучше всего рассматривать как настройки алгоритма машинного обучения, которые настраиваются специалистом по данным перед тренировкой. Примерами могут служить количество деревьев в случайном лесу или количество соседей, используемых в регрессии K ближайших соседей.
Параметры модели - это то, что модель изучает во время обучения, например, веса в линейной регрессии.
Мы, как исследователи данных, контролируем модель, выбирая гиперпараметры, и этот выбор может оказать существенное влияние на конечную производительность модели (хотя, как правило, не так сильно, как получение большего количества данных или технических характеристик).

Настройка гиперпараметров модели управляет балансом по сравнению с подгонкой в ​​модели. Мы можем попытаться исправить недооценку, сделав более сложную модель, например, используя больше деревьев в случайном лесу или больше слоев в глубокой нейронной сети. Модель, которая недостаточно подходит, имеет высокий уклон и возникает, когда нашей модели не хватает емкости (степени свободы) для изучения взаимосвязи между функциями и целью. Мы можем попытаться исправить переоснащение, ограничивая сложность модели и применяя регуляризацию. Это может означать уменьшение степени полиномиальной регрессии или добавление выпадающих слоев в глубокую нейронную сеть. Модель, которая переоснащается, имеет высокую дисперсию и фактически запоминает тренировочный набор. Как недостаточное, так и дополнительное оснащение приводят к плохой производительности обобщения в тестовом наборе.

Проблема с выбором гиперпараметров состоит в том, что ни один набор не будет работать лучше всех проблем. Поэтому для каждого нового набора данных мы должны найти лучшие настройки. Это может занять много времени, но, к счастью, в Scikit-Learn есть несколько вариантов выполнения этой процедуры. Более того, новые библиотеки, такие как TPOT от epistasis labs, стремятся сделать этот процесс автоматически для вас! Пока что мы будем делать это вручную (вроде) в Scikit-Learn, но следите за статьями об автоматическом выборе модели!

### Hyperparameter Tuning with Random Search and Cross Validation

```python
loss = ['ls', 'lad', 'huber']

# Number of trees used in the boosting process
n_estimators = [100, 500, 900, 1100, 1500]

# Maximum depth of each tree
max_depth = [2, 3, 5, 10, 15]

# Minimum number of samples per leaf
min_samples_leaf = [1, 2, 4, 6, 8]

# Minimum number of samples to split a node
min_samples_split = [2, 4, 6, 10]

# Maximum number of features to consider for making splits
max_features = ['auto', 'sqrt', 'log2', None]

# Define the grid of hyperparameters to search
hyperparameter_grid = {'loss': loss,
                       'n_estimators': n_estimators,
                       'max_depth': max_depth,
                       'min_samples_leaf': min_samples_leaf,
                       'min_samples_split': min_samples_split,
                       'max_features': max_features}
```

Мы выбрали 6 различных гиперпараметров, чтобы настроить регрессор повышения градиента. Все это будет влиять на модель различными способами, которые трудно определить заранее, и единственный способ найти лучшую комбинацию для конкретной проблемы - это проверить их! Чтобы прочитать о гиперпараметрах, я предлагаю взглянуть на документацию Scikit-Learn. На данный момент, просто знайте, что мы пытаемся найти лучшую комбинацию гиперпараметров, и поскольку нет теории, которая могла бы сказать нам, что будет работать лучше всего, мы просто должны оценить их, как запустить эксперимент!

В приведенном ниже коде мы создаем объект рандомизированного поиска, передавая следующие параметры:

оценка: модель
param_distributions: распределение параметров, которые мы определили

cv количество сгибов, используемых для перекрестной проверки k-кратных

n_iter: количество различных комбинаций, которые нужно попробовать
оценка: какой показатель использовать при оценке кандидатов

n_jobs: количество ядер для параллельной работы (-1 будет использовать все доступные)

verbose: сколько информации отображать (1 отображает ограниченное количество)

return_train_score: вернуть результат обучения для каждого раза перекрестной проверки

random_state: исправляет используемый генератор случайных чисел, поэтому мы получаем одинаковые результаты при каждом запуске
Рандомизированный объект поиска обучается так же, как и любая другая модель 
scikit-learn. После тренировки мы можем сравнить все комбинации гиперпараметров и найти наиболее эффективную комбинацию.

```python
# Create the model to use for hyperparameter tuning
model = GradientBoostingRegressor(random_state = 42)

# Set up the random search with 4-fold cross validation
random_cv = RandomizedSearchCV(estimator=model,
                               param_distributions=hyperparameter_grid,
                               cv=4, n_iter=25, 
                               scoring = 'neg_mean_absolute_error',
                               n_jobs = -1, verbose = 1, 
                               return_train_score = True,
                               random_state=42)
```

```python
random_cv.fit(X, y)
```

Scikit-learn использует абсолютную ошибку отрицательного среднего для оценки, потому что хочет максимизировать показатель. Поэтому лучший результат будет ближе к 0. Мы можем получить результаты рандомизированного поиска в кадре данных и отсортировать значения по производительности.

```python
# Get all of the cv results and sort by the test performance Получить все результаты и отсортировать результаты теста
random_results = pd.DataFrame(random_cv.cv_results_).sort_values('mean_test_score', ascending = False)

random_results.head()
```

```python
random_cv.best_estimator_
```

Лучшая модель с градиентным усилением имеет следующие гиперпараметры:

        loss = lad

        n_estimators = 500

        max_depth = 5

        min_samples_leaf = 6

        min_samples_split = 6

        max_features = None (это означает, что max_features = n_features в соответствии с документами)

Использование случайного поиска - хороший способ сузить возможные гиперпараметры, которые нужно попробовать. Изначально мы не знали, какая комбинация будет работать лучше, но это, по крайней мере, сужает диапазон вариантов.

Мы могли бы использовать результаты случайного поиска для информирования поиска по сетке, создавая сетку с гиперпараметрами, близкими к тем, которые работали лучше всего при рандомизированном поиске. Однако вместо того, чтобы снова оценивать все эти параметры, я остановлюсь на одном - количестве деревьев в лесу (n_estimators). Изменяя только один гиперпараметр, мы можем непосредственно наблюдать, как он влияет на производительность. В случае количества деревьев мы ожидаем увидеть существенное влияние на количество недопоставок.

Здесь мы будем использовать поиск по сетке с сеткой, которая имеет только гиперпараметр n_estimators. Мы оценим диапазон деревьев, затем построим график тренировок и результатов тестирования, чтобы понять, как увеличение количества деревьев влияет на нашу модель. Мы исправим другие гиперпараметры в лучших значениях, возвращаемых при случайном поиске, чтобы изолировать эффект количества деревьев.

```python
# Create a range of trees to evaluate
trees_grid = {'n_estimators': [100, 150, 200, 250, 300, 350, 400, 450, 500, 550, 600, 650, 700, 750, 800]}

model = GradientBoostingRegressor(loss = 'lad', max_depth = 5,
                                  min_samples_leaf = 6,
                                  min_samples_split = 6,
                                  max_features = None,
                                  random_state = 42)

# Grid Search Object using the trees range and the random forest model
grid_search = GridSearchCV(estimator = model, param_grid=trees_grid, cv = 4, 
                           scoring = 'neg_mean_absolute_error', verbose = 1,
                           n_jobs = -1, return_train_score = True)
```

```python
# Fit the grid search
grid_search.fit(X, y)
```

```python
# Get the results into a dataframe
results = pd.DataFrame(grid_search.cv_results_)

# Plot the training and testing error vs number of trees
figsize(8, 8)
plt.style.use('fivethirtyeight')
plt.plot(results['param_n_estimators'], -1 * results['mean_test_score'], label = 'Testing Error')
plt.plot(results['param_n_estimators'], -1 * results['mean_train_score'], label = 'Training Error')
plt.xlabel('Number of Trees'); plt.ylabel('Mean Abosolute Error'); plt.legend();
plt.title('Performance vs Number of Trees');
```

```python
results.sort_values('mean_test_score', ascending = False).head(5)
```

Из этого сюжета видно, что наша модель переоснащается! Ошибка обучения значительно ниже, чем ошибка тестирования, что показывает, что модель очень хорошо изучает данные обучения, но затем не может обобщать и данные теста. Перемещение, поскольку количество деревьев увеличивается, количество переоснащения увеличивается. И тест, и ошибка обучения уменьшаются с увеличением количества деревьев, но ошибка обучения уменьшается быстрее.

Всегда будет разница между ошибкой обучения и ошибкой тестирования (ошибка обучения всегда меньше), но если есть существенная разница, мы хотим попытаться уменьшить переоснащение, либо получая больше данных обучения, либо уменьшая сложность модели. через настройку гиперпараметра или регуляризацию. Для регрессора повышения градиента некоторые опции включают уменьшение количества деревьев, уменьшение максимальной глубины каждого дерева и увеличение минимального количества выборок в листовом узле. Для тех, кто хочет углубиться в регрессор повышения градиента, вот отличная статья. На данный момент мы будем использовать модель с наилучшими характеристиками и допустим, что она может соответствовать учебному набору.

Основываясь на результатах перекрестной проверки, лучшая модель использует 800 деревьев и достигает ошибки перекрестной проверки ниже 9. Это указывает на то, что средняя оценка перекрестной проверки для показателя Energy Star находится в пределах 9 баллов от истинного ответа!

### Оценить окончательную модель на тестовом наборе

Мы будем использовать лучшую модель из настройки гиперпараметра, чтобы делать прогнозы на тестовом наборе. Помните, что наша модель никогда не видела тестовый набор раньше, поэтому эта производительность должна быть хорошим показателем того, как модель будет работать, если ее развернуть в реальном мире.

Для сравнения мы также можем посмотреть на производительность модели по умолчанию. Приведенный ниже код создает окончательную модель, обучает ее (с учетом времени) и оценивает набор тестов.

```python
# Default model
default_model = GradientBoostingRegressor(random_state = 42)

# Select the best model
final_model = grid_search.best_estimator_

final_model
```

```python

%%timeit -n 1 -r 5
default_model.fit(X, y)
```

```python
%%timeit -n 1 -r 5
final_model.fit(X, y)
```

```python
#Производительность модели по умолчанию на тестовом наборе:

default_pred = default_model.predict(X_test)
final_pred = final_model.predict(X_test)

print('Default model performance on the test set: MAE = %0.4f.' % mae(y_test, default_pred))
print('Final model performance on the test set:   MAE = %0.4f.' % mae(y_test, final_pred))
```

Последняя модель превосходит базовую модель примерно на 10%, но за счет значительного увеличения времени работы (она примерно в 12 раз медленнее на моей машине). Машинное обучение часто является областью компромиссов: смещение против дисперсии, точность против интерпретируемости, точность против времени выполнения, и окончательное решение о том, какую модель использовать, зависит от ситуации. Здесь увеличение времени выполнения не является препятствием, поскольку, хотя относительная разница велика, абсолютная величина времени тренировки не является значительной. В другой ситуации баланс может не совпадать, поэтому нам нужно учитывать то, что мы оптимизируем, и ограничения, с которыми нам приходится работать.

Чтобы получить представление о прогнозах, мы можем построить распределение истинных значений в тестовом наборе и прогнозных значений в тестовом наборе.

```python
sns.kdeplot
```

Распределение выглядит почти одинаковым, хотя плотность прогнозируемых значений ближе к медиане тестовых значений, чем к фактическому пику в 100. Кажется, что модель может быть менее точной при прогнозировании экстремальных значений и вместо этого прогнозирует значения ближе к медиане.

Другой диагностический график - это гистограмма остатков. В идеале, мы надеемся, что остатки обычно распределяются, а это означает, что модель ошибочна в одинаковом количестве в обоих направлениях (высокий и низкий).

```python
figsize = (6, 6)

# Calculate the residuals 
residuals = final_pred - y_test

# Plot the residuals in a histogram
plt.hist(residuals, color = 'red', bins = 20,
         edgecolor = 'black')
plt.xlabel('Error'); plt.ylabel('Count')
plt.title('Distribution of Residuals');
```

Остатки близки к нормально распределенным, с несколькими заметными выбросами на нижнем конце. Это указывает на ошибки, когда оценка модели была намного ниже, чем истинное значение.
