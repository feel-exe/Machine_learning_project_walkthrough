### Пропуски
Теперь, когда у нас есть правильные типы данных столбцов, мы можем начать анализ, посмотрев на процент пропущенных значений в каждом столбце. Пропущенные значения хороши, когда мы проводим анализ поисковых данных, но они должны быть заполнены для методов машинного обучения.

Ниже приведена функция, которая вычисляет количество пропущенных значений и процент от общих значений, пропущенных для каждого столбца. Как и во многих задачах в науке о данных, я сам не писал эту функцию, а нашел ее в Интернете на форуме Stack Overflow!

```python
def missing_values_table(df):
        # Total missing values
        mis_val = df.isnull().sum()
        
        # Percentage of missing values
        mis_val_percent = 100 * df.isnull().sum() / len(df)
        
        # Make a table with the results
        mis_val_table = pd.concat([mis_val, mis_val_percent], axis=1)
        
        # Rename the columns
        mis_val_table_ren_columns = mis_val_table.rename(
        columns = {0 : 'Missing Values', 1 : '% of Total Values'})
        
        # Sort the table by percentage of missing descending
        mis_val_table_ren_columns = mis_val_table_ren_columns[
            mis_val_table_ren_columns.iloc[:,1] != 0].sort_values(
        '% of Total Values', ascending=False).round(1)
        
        # Print some summary information
        print ("Your selected dataframe has " + str(df.shape[1]) + " columns.\n"      
            "There are " + str(mis_val_table_ren_columns.shape[0]) +
              " columns that have missing values.")
        
        # Return the dataframe with missing information
        return mis_val_table_ren_columns
```

```python
missing_values_table(data)
```

Какие столбцы для сохранения могут быть немного произвольными (здесь обсуждение), но для этого проекта мы удалим все столбцы с пропущенными значениями более чем на 50%. В общем, будьте осторожны с отбрасыванием любой информации, потому что даже если ее нет для всех наблюдений, она все равно может быть полезна для прогнозирования целевого значения.

```python
# Get the columns with > 50% missing
missing_df = missing_values_table(data);
missing_columns = list(missing_df[missing_df['% of Total Values'] > 50].index)
print('We will remove %d columns.' % len(missing_columns))
```

```python
# Drop the columns
data = data.drop(columns = list(missing_columns))

# For older versions of pandas (https://github.com/pandas-dev/pandas/issues/19078)
# data = data.drop(list(missing_columns), axis = 1)
```

## Удаление выбросов

Когда мы удаляем выбросы, мы хотим быть осторожными, чтобы не выбрасывать измерения только потому, что они выглядят странно. Они могут быть результатом фактического явления, которое мы должны дополнительно исследовать. При удалении выбросов я стараюсь быть максимально консервативным, используя определение экстремального выброса:

На нижнем уровне экстремальный выброс ниже $ \ text {First Quartile} -3 * \ text {Interquartile Range} $
На верхнем уровне экстремальный выброс выше $ \ text {Third Quartile} + 3 * \ text {Interquartile Range} $
В этом случае я удалю только одну отдаленную точку и посмотрю, как выглядит дистрибутив.

```python

# Calculate first and third quartile
first_quartile = data['Site EUI (kBtu/ft²)'].describe()['25%']
third_quartile = data['Site EUI (kBtu/ft²)'].describe()['75%']

# Interquartile range
iqr = third_quartile - first_quartile
print('квантиль', iqr)
# Remove outliers
data = data[(data['Site EUI (kBtu/ft²)'] > (first_quartile - 3 * iqr)) &
            (data['Site EUI (kBtu/ft²)'] < (third_quartile + 3 * iqr))]
```

```python
# Histogram Plot of Site EUI
figsize(8, 8)
plt.hist(data['Site EUI (kBtu/ft²)'].dropna(), bins = 20, edgecolor = 'black');
plt.xlabel('Site EUI'); 
plt.ylabel('Count'); plt.title('Site EUI Distribution');
```

## Корреляции между объектами и целью

Чтобы количественно определить корреляции между признаками (переменными) и целью, мы можем рассчитать коэффициент корреляции Пирсона. Это мера силы и направления линейного отношения между двумя переменными: значение -1 означает, что две переменные совершенно отрицательно линейно коррелированы, а значение +1 означает, что две переменные совершенно положительно линейно коррелированы. На рисунке ниже показаны различные значения коэффициента корреляции и их графическое отображение.

Хотя могут быть нелинейные отношения между объектами и целями, и коэффициенты корреляции не учитывают взаимодействия между объектами, линейные отношения являются хорошим способом начать изучение тенденций в данных. Затем мы можем использовать эти значения для выбора функций для использования в нашей модели.

Код ниже вычисляет коэффициенты корреляции между всеми переменными и оценкой.

```python
# Find all correlations and sort 
correlations_data = data.corr()['score'].sort_values()

# Print the most negative correlations
print(correlations_data.head(15), '\n')

# Print the most positive correlations
print(correlations_data.tail(15))
```

Есть несколько сильных отрицательных корреляций между особенностями и целью. Наиболее отрицательные корреляции с оценкой - это различные категории интенсивности использования энергии (EUI), EUI для площадки (кБту / фут²) и EUI для нормализованной погоды (кБту / фут²) (они незначительно отличаются по способу расчета). EUI - это количество энергии, потребляемой зданием, деленное на квадратные метры зданий и предназначенное для измерения эффективности здания с более низким баллом. Интуитивно понятно, что эти корреляции имеют смысл: по мере увеличения EUI показатель Energy Star имеет тенденцию к снижению.

Чтобы учесть возможные нелинейные отношения, мы можем взять квадратные корни и натуральные логарифмические преобразования объектов, а затем вычислить коэффициенты корреляции с оценкой. Чтобы попытаться отразить любые возможные отношения между районом или типом здания (помните, что это категориальные переменные) и оценкой, мы должны будем в горячем виде кодировать эти столбцы.

В следующем коде мы берем логарифмические преобразования и преобразования квадратного корня числовых переменных, однозначно кодируем две выбранные категориальные переменные (тип здания и район), вычисляем корреляции между всеми функциями и оценкой и отображаем верхние 15 самые положительные и 15 самых отрицательных корреляций. Это много, но с пандами это просто сделать каждый шаг!

```python
# Оценка взаимной корреляции
def r2(df):
        
        df_pd_corr = df.corr()
        columns = list(df_pd_corr.columns)
        #columns.remove(attribute_column) 
        df_corr = pd.DataFrame(columns = ['first', 'second', 'r2'])
        
        k = 0
        for h in range(len(df_pd_corr)):
            for g in range(h):
                df_corr.loc[k, 'first'] = columns[h]
                df_corr.loc[k, 'second'] = columns[g]
                df_corr.loc[k, 'r2'] = df_pd_corr.iloc[h, g]
                #print(k)
                #print(df_pd_corr.iloc[h, g])
                k += 1
        
        
        df_corr = df_corr[
            df_corr.iloc[:,1] != 0].sort_values(
        'r2', ascending=False).round(1)
        
        return df_corr
```

```python
# Select the numeric columns
numeric_subset = data.select_dtypes('number')

# Create columns with square root and log of numeric columns
for col in numeric_subset.columns:
    # Skip the Energy Star Score column
    if col == 'score':
        next
    else:
        numeric_subset['sqrt_' + col] = np.sqrt(numeric_subset[col])
        numeric_subset['log_' + col] = np.log(numeric_subset[col])

# Select the categorical columns
categorical_subset = data[['Borough', 'Largest Property Use Type']]

# One hot encode
categorical_subset = pd.get_dummies(categorical_subset)

# Join the two dataframes using concat
# Make sure to use axis = 1 to perform a column bind
features = pd.concat([numeric_subset, categorical_subset], axis = 1)

# Drop buildings without an energy star score
features = features.dropna(subset = ['score'])

# Find correlations with the score 
correlations = features.corr()['score'].dropna().sort_values()
```

## Pairs Plot

В качестве заключительного упражнения для исследовательского анализа данных мы можем составить график пар между несколькими различными переменными. График пар - это отличный способ исследовать множество переменных одновременно, поскольку он показывает диаграммы рассеяния между парами переменных и гистограммами отдельных переменных на диагонали.

Используя функцию seaborn PairGrid, мы можем сопоставить различные графики с тремя аспектами сетки. Верхний треугольник будет иметь диаграммы рассеяния, диагональ покажет гистограммы, а нижний треугольник покажет как коэффициент корреляции между двумя переменными, так и двухмерную оценку плотности ядра двух переменных.

```python

# Extract the columns to  plot
plot_data = features[['score', 'Site EUI (kBtu/ft²)', 
                      'Weather Normalized Source EUI (kBtu/ft²)', 
                      'log_Total GHG Emissions (Metric Tons CO2e)']]

# Replace the inf with nan
plot_data = plot_data.replace({np.inf: np.nan, -np.inf: np.nan})

# Rename columns 
plot_data = plot_data.rename(columns = {'Site EUI (kBtu/ft²)': 'Site EUI', 
                                        'Weather Normalized Source EUI (kBtu/ft²)': 'Weather Norm EUI',
                                        'log_Total GHG Emissions (Metric Tons CO2e)': 'log GHG Emissions'})

# Drop na values
plot_data = plot_data.dropna()

# Function to calculate correlation coefficient between two columns
def corr_func(x, y, **kwargs):
    r = np.corrcoef(x, y)[0][1]
    ax = plt.gca()
    ax.annotate("r = {:.2f}".format(r),
                xy=(.2, .8), xycoords=ax.transAxes,
                size = 20)

# Create the pairgrid object
grid = sns.PairGrid(data = plot_data, size = 3)

# Upper is a scatter plot
grid.map_upper(plt.scatter, color = 'red', alpha = 0.6)

# Diagonal is a histogram
grid.map_diag(plt.hist, color = 'red', edgecolor = 'black')

# Bottom is correlation and density plot
grid.map_lower(corr_func);
grid.map_lower(sns.kdeplot, cmap = plt.cm.Reds)

# Title for entire plot
plt.suptitle('Pairs Plot of Energy Data', size = 36, y = 1.02);
```

Чтобы интерпретировать отношения на графике, мы можем искать, где переменные в одной строке пересекаются с переменными в одном столбце. Например, чтобы найти связь между оценкой и журналом выбросов парниковых газов, мы смотрим на столбец оценки и находим строку журнала выбросов парниковых газов. На пересечении (нижний левый график) мы видим, что показатель имеет коэффициент корреляции -0,35 с этой переменной. Если мы посмотрим на верхний правый график, мы можем увидеть график рассеяния этих отношений.

Он также печатает корреляции, которые он удаляет, чтобы мы могли видеть эффект корректировки порога. Мы будем использовать порог 0,6, который удаляет одну из пары объектов, если коэффициент корреляции между объектами превышает это значение.

```python
def remove_collinear_features(x, threshold):
    '''
    Objective:
        Remove collinear features in a dataframe with a correlation coefficient
        greater than the threshold. Removing collinear features can help a model
        to generalize and improves the interpretability of the model.
        
    Inputs: 
        threshold: any features with correlations greater than this value are removed
    
    Output: 
        dataframe that contains only the non-highly-collinear features
    '''
    
    # Dont want to remove correlations between Energy Star Score
    y = x['score']
    x = x.drop(columns = ['score'])
    
    # Calculate the correlation matrix
    corr_matrix = x.corr()
    iters = range(len(corr_matrix.columns) - 1)
    drop_cols = []

    # Iterate through the correlation matrix and compare correlations
    for i in iters:
        for j in range(i):
            item = corr_matrix.iloc[j:(j+1), (i+1):(i+2)]
            col = item.columns
            row = item.index
            val = abs(item.values)
            
            # If correlation exceeds the threshold
            if val >= threshold:
                # Print the correlated features and the correlation value
                # print(col.values[0], "|", row.values[0], "|", round(val[0][0], 2))
                drop_cols.append(col.values[0])

    # Drop one of each pair of correlated columns
    drops = set(drop_cols)
    x = x.drop(columns = drops)
    x = x.drop(columns = ['Weather Normalized Site EUI (kBtu/ft²)', 
                          'Water Use (All Water Sources) (kgal)',
                          'log_Water Use (All Water Sources) (kgal)',
                          'Largest Property Use Type - Gross Floor Area (ft²)'])
    
    # Add the score back in to the data
    x['score'] = y
               
    return x
```

```python

# Remove the collinear features above a specified correlation coefficient
features = remove_collinear_features(features, 0.6);
```

```python
# Remove any columns with all na values
features  = features.dropna(axis=1, how = 'all')
features.shape
```

Это все еще довольно мало, но в основном это потому, что мы однозначно закодировали категориальные переменные. Кроме того, в то время как большое количество признаков может быть проблематичным для моделей, таких как линейная регрессия, модели, такие как случайный лес, выполняют неявный выбор признаков и автоматически определяют, какие признаки важны во время обучения. Есть и другие шаги по выбору функций, но сейчас мы сохраним все имеющиеся у нас функции и посмотрим, как работает модель.

Выбор дополнительных функций
Существует множество других способов выбора функций. Некоторые популярные методы включают анализ основных компонентов (PCA), который преобразует функции в уменьшенное число измерений, которые сохраняют наибольшую дисперсию, или анализ независимых компонентов (ICA), который направлен на поиск независимых источников в наборе функций. Однако, хотя эти методы эффективны для уменьшения числа признаков, они создают новые функции, которые не имеют физического смысла и, следовательно, делают интерпретацию модели практически невозможной.

Эти методы очень полезны для работы с многомерными данными, и я бы посоветовал прочитать больше на эту тему, если вы планируете решать проблемы машинного обучения!

# Разделить на учебные и тестовые наборы

В машинном обучении нам всегда нужно разделять наши функции на два набора:

Учебный набор, который мы предоставляем нашей модели во время обучения вместе с ответами, чтобы он мог изучить соответствие между функциями и целью.
Набор для тестирования, который мы используем для оценки сопоставления, полученного моделью. Модель никогда не видела ответов на тестовом наборе, но вместо этого должна делать прогнозы, используя только функции. Поскольку мы знаем истинные ответы для тестового набора, мы можем затем сравнить прогнозы теста с истинными целями теста, чтобы получить оценку того, насколько хорошо наша модель будет работать при развертывании в реальном мире.
Для нашей задачи мы сначала извлечем все здания без оценки Energy Star (мы не знаем верного ответа для этих зданий, поэтому они не будут полезны для обучения или тестирования). Затем мы разделим здания со счетом Energy Star на набор для тестирования 30% зданий и тренировочный набор для 70% зданий.

Разбить данные на случайные обучающие и тестовые наборы просто с помощью scikit-learn. Мы можем установить случайное состояние разделения, чтобы обеспечить согласованные результаты.

```python

# Extract the buildings with no score and the buildings with a score
no_score = features[features['score'].isna()]
score = features[features['score'].notnull()]

print(no_score.shape)
print(score.shape)
```

```python
# Separate out the features and targets
features = score.drop(columns='score')
targets = pd.DataFrame(score['score'])

# Replace the inf and -inf with nan (required for later imputation)
features = features.replace({np.inf: np.nan, -np.inf: np.nan})

# Split into 70% training and 30% testing set
X, X_test, y, y_test = train_test_split(features, targets, test_size = 0.3, random_state = 42)

print(X.shape)
print(X_test.shape)
print(y.shape)
print(y_test.shape)
```

У нас есть 1858 зданий без оценки, 6622 здания с оценкой в наборе обучения и 2839 зданий с оценкой в наборе тестирования. У нас есть один последний шаг в этой записной книжке: определение наивного базового уровня для наших моделей.

Установить базовый уровень
Важно установить наивный базовый уровень, прежде чем мы начнем создавать модели машинного обучения. Если модели, которые мы создаем, не могут превзойти наивное предположение, то нам, возможно, придется признать, что машинное обучение не подходит для этой проблемы. Это может быть потому, что мы не используем правильные модели, потому что нам нужно больше данных, или потому что есть более простое решение, которое не требует машинного обучения. Установление базового уровня имеет решающее значение, поэтому мы не заканчиваем построением модели машинного обучения только для того, чтобы понять, что на самом деле не можем решить проблему.

Для задачи регрессии хорошим наивным исходным уровнем является прогнозирование медианного значения цели в тренировочном наборе для всех примеров в тестовом наборе. Это просто реализовать и устанавливает относительно низкий уровень для наших моделей: если они не могут лучше, чем угадывать значение медин, нам нужно будет переосмыслить наш подход.

Метрика: средняя абсолютная ошибка
Есть несколько показателей, используемых в задачах машинного обучения, и может быть трудно определить, какой из них выбрать. В большинстве случаев это будет зависеть от конкретной проблемы и от того, для чего вам нужна оптимизация. Мне нравится совет Эндрю Нга использовать одну метрику производительности в реальном выражении для сравнения моделей, потому что это упрощает процесс оценки. Вместо того, чтобы вычислять несколько метрик и пытаться определить, насколько важен каждый из них, мы должны использовать одно число. В этом случае, поскольку мы делаем регрессию, средняя абсолютная ошибка является подходящей метрикой. Это также интерпретируемо, потому что представляет среднюю сумму, которую мы оцениваем, если она не совпадает с целевым значением.

Функция ниже вычисляет среднюю абсолютную ошибку между истинными значениями и прогнозами.

```python
# Function to calculate mean absolute error
def mae(y_true, y_pred):
    return np.mean(abs(y_true - y_pred))
```

```python
baseline_guess = np.median(y)

print('The baseline guess is a score of %0.2f' % baseline_guess)
print("Baseline Performance on the test set: MAE = %0.4f" % mae(y_test, baseline_guess))
```

Это показывает, что наша средняя оценка на тестовом наборе отключена примерно на 25 баллов. Баллы находятся в диапазоне от 1 до 100, так что это означает среднюю ошибку от наивного метода, если она составляет около 25%. Наивный метод угадывания медианного значения тренировки дает нам низкую базовую оценку для наших моделей, чтобы победить!
